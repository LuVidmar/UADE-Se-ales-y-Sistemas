### Ejercicio 3
Desarrolle un de scripts en el lenguaje de programación *Python*, el cual describa y grafique una señal de potencia.

Onda cuadrada:
$
\\
x_{(t)}=\displaystyle \sum_{i=-\infty}^{\infty}
\Pi_{(t-i)}
$

Calculo de potencia:
$
\\
P=\displaystyle
\lim_{T\to \infty}
\frac{1}{2T}
\int_{-T}^T
|x_{(t)}|^2
dt
$
$
\displaystyle
=\lim_{T\to \infty}
\frac{1}{2T}
\int_{-T}^T
|
\sum_{i=-\infty}^{\infty}
\Pi_{(t-i)}
|^2
dt
$

- La señal es positiva en todo el dominio, se puede sacar el modulo
- Se divide la integral entre las partes donde la señal es cero y donde es uno y luego se utiliza el principio de superposicion

Cuando la señal es 1:
$
\\
\displaystyle
P_1
=\lim_{T\to \infty}
\frac{1}{2T}
\int_{-T}^T
{
\sum_{i=-\infty}^{\infty}
1
}
^2
dt
=\lim_{T\to \infty}
\frac{1}{2T}
\int_{-T}^T
1
^2
dt
=\lim_{T\to \infty}
\frac{1}{2T}
\to0
$

Cuando la señal es 2:
$
\\
\displaystyle
P_1
=\lim_{T\to \infty}
\frac{1}{2T}
\int_{-T}^T
{
\sum_{i=-\infty}^{\infty}
0
}
^2
dt
=0
$

$\therefore P=P_1+P_2
\\
0<P<\infty \implies$ es señal de potencia
